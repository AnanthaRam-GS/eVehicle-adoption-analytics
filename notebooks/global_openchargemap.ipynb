```python
import requests
import pandas as pd
import time

API_KEY = "87de7775-b1f5-4bde-ad0c-925efebdcda9"

url = "https://api.openchargemap.io/v3/poi/"

# Major EV countries
countries = [
    "US",  # United States
    "CN",  # China
    "DE",  # Germany
    "GB",  # United Kingdom
    "NO",  # Norway
    "NL",  # Netherlands
    "FR",  # France
    "IN",  # India
    "CA",  # Canada
    "JP"   # Japan
]

all_data = []

for country in countries:
    
    print(f"Downloading data for {country}...")
    
    params = {
        "output": "json",
        "countrycode": country,
        "maxresults": 5000,
        "compact": True,
        "verbose": False,
        "key": API_KEY
    }
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        
        df = pd.json_normalize(data)
        df["country_code"] = country
        
        all_data.append(df)
        
    else:
        print(f"Failed for {country}")
    
    time.sleep(1)  # avoid API rate limits


# Combine all countries
charging_df = pd.concat(all_data, ignore_index=True)

print("Total records:", charging_df.shape)

# Save raw dataset
charging_df.to_csv("../data/raw/openchargemap_global_raw.csv", index=False)

print("OpenChargeMap dataset saved successfully.")
```
